```python
import folium
import pandas as pd
latitude = 22.6
longitude = 120.3

cdata = pd.read_csv('https://asdkmm5050.github.io/data/1090226.csv')
incidents = folium.map.FeatureGroup()
for lat, lng, in zip(cdata['緯度'], cdata['經度']):
  incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )
map = folium.Map(location=[latitude, longitude], zoom_start=12)
map.add_child(incidents)

```
